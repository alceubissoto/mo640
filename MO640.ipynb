# Algo 3

```python
## Algo 3

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random 
df = pd.read_csv('results_algo3.csv', names=['iteration', 'timestamp','fitness','dataset','matrix','algo'])

files = df['matrix'].unique()
for f in random.sample(list(files), 25):
    df_matrix = df[df['matrix']==f]
    plt.title(f)
    plt.xlabel('iteration')
    plt.ylabel('fitness')
    plt.plot(df_matrix['iteration'], df_matrix['fitness'], 'ro')
    plt.show()
```

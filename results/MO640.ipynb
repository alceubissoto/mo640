# Overall results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random 
```

```python
def overall_results(file):
    print('Results for ', file)
    df = pd.read_csv(file, names=['iteration', 'timestamp','fitness','dataset','matrix','algo'])

    starts = df.groupby('matrix').first()['fitness']
    ends = df.groupby('matrix').last()['fitness']
    x = (starts-ends)/starts*100

    # the histogram of the data
    plt.hist(x, 50, facecolor='green', alpha=0.75)
    plt.xlabel('Reduction')
    plt.ylabel('# of samples')
    plt.title(file + '- Histogram of reductions in %')
    plt.grid(True)

    plt.show()
    print(np.histogram(x))

overall_results('results_algo1.csv')
overall_results('results_algo2.csv')
overall_results('results_algo3.csv')
overall_results('results_algo4.csv')
overall_results('results_algo5.csv')
```

# Algo 3

```python
## Algo 3

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random 
df = pd.read_csv('results_algo3.csv', names=['iteration', 'timestamp','fitness','dataset','matrix','algo'])

files = df['matrix'].unique()
for f in random.sample(list(files), 25):
    df_matrix = df[df['matrix']==f]
    plt.title(f)
    plt.xlabel('iteration')
    plt.ylabel('fitness')
    plt.plot(df_matrix['iteration'], df_matrix['fitness'], 'ro')
    plt.show()
```
